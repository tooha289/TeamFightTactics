## [이슈 항목]

### 1. 리스트에 다른 리스트의 원소를 추가할 때 append를 사용




```
list1 = [1,2,3]
list2 = [4,5,6]
```

두 리스트가 존재 할 때 아래와 같이 사용하여 예상치 못한 부분에서 에러가 발생했었습니다.

```
list1.append(list2) # [1,2,3,[4,5,6]]
```

위의 결과를 원했던 것이 아니어서 아래와 같이 수정했습니다.

```
list1.extend(list2) # [1,2,3,4,5,6]
list1 + list2 # [1,2,3,4,5,6]

```
